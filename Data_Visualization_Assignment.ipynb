```python
# Q.5  Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

## Ans:

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)

# Create a covariance matrix with varying degrees of correlation
cov_matrix = np.array([[1, 0.8, 0.5],
                       [0.8, 1, 0.3],
                       [0.5, 0.3, 1]])

data = np.random.multivariate_normal(mean=[0, 0, 0], cov=cov_matrix, size=1000)

df = pd.DataFrame(data, columns=["Feature 1", "Feature 2", "Feature 3"])

print(df.head())

# Visualize the correlation matrix using a heatmap
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap="viridis")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Q.1 Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

## Ans:

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')
p.line(x, y, legend_label="sin(x)", line_width=2)

show(p)
```

```python
# Q.2 Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

## Ans:

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

output_notebook()

N = 100
x = np.random.rand(N)
y = np.random.rand(N)
sizes = np.random.randint(5, 30, N)
colors = np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], N)

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

p = figure(title="Random Scatter Plot", x_axis_label='x', y_axis_label='y')
p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_field='colors')

show(p)
```

```python
# Q.3 Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

## Ans:

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts')

p.vbar(x='fruits', top='counts', width=0.8, source=source, color="skyblue")

show(p)
```

```python
# Q.4  Create a Bokeh histogram to visualize the distribution of the given data.
        # data_hist = np.random.randn(1000)
        # hist, edges = np.histogram(data_hist, bins=30)

## Ans:

import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

output_notebook()

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

source = ColumnDataSource(data=dict(top=hist, left=edges[:-1], right=edges[1:]))

p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')

p.quad(top='top', bottom=0, left='left', right='right', source=source, fill_color='navy', line_color='white')

show(p)
```

```python
# Q.5  Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

## Ans:

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.palettes import Viridis256

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

data = {'x': xx.flatten(), 'y': yy.flatten(), 'value': data_heatmap.flatten()}

output_notebook()

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y')
mapper = linear_cmap(field_name='value', palette= Viridis256, low=data['value'].min(), high=data['value'].max())

p.rect(x='x', y='y', width=0.1, height=0.1, source=data, fill_color=mapper)

color_bar = ColorBar(color_mapper=mapper['transform'], location=(0,0))
p.add_layout(color_bar, 'right')

show(p)
```
